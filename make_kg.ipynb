```python
nodes = {
    'asset': {},
    'composer': {}
}
nodes_remap = {}
idx = 0
for i in rdf:
    s, p , o = i
    splitted_uri = str(s).split('/')
    if splitted_uri[-2] == 'asset':
        nodes['asset'][splitted_uri[-1]] = {}
    elif splitted_uri[-2] == 'composer':
        nodes['composer'][splitted_uri[-1]] = {}    

    if str(o).split('/')[-1] not in nodes_remap.keys():
        nodes_remap[str(o).split('/')[-1]] = idx
        idx += 1      
    
```

```python
relation = {
    'http://www.w3.org/ns/dcat#inseries': 0,
    'http://purl.org/dc/terms/creator': 1,
    'https://www.kw.ac.kr/k-terms/death': 2,
    'http://www.w3.org/ns/dcat#last': 3,
    'http://www.w3.org/1999/02/22-rdf-syntax-ns#type': 4,
    'https://www.kw.ac.kr/k-terms/movement': 5,
    'http://www.w3.org/ns/dcat#prev': 6,
    'http://www.w3.org/ns/dcat#theme': 7,
    'http://purl.org/dc/terms/title': 8,
    'https://www.kw.ac.kr/k-terms/ensemble': 9,
    'http://www.w3.org/ns/dcat#first': 10,
    'https://www.kw.ac.kr/k-terms/birth': 11
}
```

```python
for triple in rdf:
    s, p, o = triple
    splitted_uri = str(s).split('/')
    if splitted_uri[-2] in ['asset', 'composer']:
        if str(p) in relation.keys():
            nodes[splitted_uri[-2]][splitted_uri[-1]][relation[str(p)]] = nodes_remap[str(o).split('/')[-1]]
 
```

```python
import random
```

```python
temp = []

for t in nodes.keys():
    for entity, g in nodes[t].items():
        s = nodes_remap[entity]
        for p, o in g.items():
            temp.append((s, p, o))

temp = sorted(temp, key=lambda x: random.random())           
```

```python
with open('data/kg_final.txt', 'w') as f:
    for triple in temp:
        f.write(f"{triple[0]} {triple[1]} {triple[2]}\n")
```
